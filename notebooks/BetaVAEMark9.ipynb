## PCA Analysis

```python
jz0_latent_codes = np.load("/home/xzcapask/atlas_ad_hllhc/data/model_data/BetaVAEMark9/l4/b1e_1/latent_vectors/jz0_latent_codes.npy").reshape(49000,4)
```

```python
pca = PCA()
pca.fit(jz0_latent_codes)
evr = pca.explained_variance_ratio_
```

```python
plt.plot(evr)
plt.xlabel("PCA Components")
plt.ylabel("EVR (Explained Variance Ratio)")
plt.title("PCA of JZ0 Latent Codes of VAE 9 \n for 7D latent space")
plt.savefig("/home/xzcapask/atlas_ad_hllhc/notebooks/JZ0_PCA_VAE9.png")
```

# Normal Distribution?

```python
jz0_chi = np.sum(jz0_latent_codes**2, axis = 1)
```

```python
chi = np.random.chisquare(4, 49000)
jz0_chi.shape
```

```python
plt.hist(jz0_chi, bins = 1000, range = (0,100), label = "JZ0", histtype= "step")
plt.hist(chi, bins = 1000, range = (0,100), label  = "$\chi^2_7$", histtype= "step")
plt.legend()
plt.xlabel("$\chi^2$")
plt.title("Is the JZ0 Latent Distribution Normal?")
plt.savefig("/home/xzcapask/atlas_ad_hllhc/notebooks/chi_squared_vae9.png")
```
